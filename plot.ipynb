```python
import numpy as np
import matplotlib.pyplot as plt

categories = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6']
values = np.array([5, 3, 6, 5, 1, 0])

n_colors = 256
cm = plt.get_cmap('Reds', n_colors)

fig, ax = plt.subplots(subplot_kw=dict(polar=True))
bars = ax.bar(np.linspace(0, 2 * np.pi, len(values), endpoint=False), values, width=0.8, bottom=0.1)

for r, bar in zip(values, bars):
    color_index = int((r - min(values)) / (max(values) - min(values)) * (n_colors - 1))
    bar.set_facecolor(cm(color_index))

ax.set_xticks(np.linspace(0, 2 * np.pi, len(categories), endpoint=False))
ax.set_xticklabels(categories)
ax.set_yticklabels([]) 
plt.show()
```
